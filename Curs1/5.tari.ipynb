```python
import matplotlib.pyplot as plt
import csv
```

```python
countries = []
population = []

with open("tari.csv") as freader:
    csv_reader = csv.DictReader(freader)
    print(csv_reader)
    
    for row in csv_reader:
        countries. append(row["country"])
        population.append(int(row["population"]))
print("Countries", countries)
print("Population", population)
```

```python
plt.title("Populatia celor mai numeroase 10 tari ale lumii")
plt.pie(population, labels=countries);
plt.savefig("10tari.png")
```
