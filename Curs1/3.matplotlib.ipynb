```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
fig.suptitle("Hello")
plt.show()
```

```python
fig = plt.figure()
fig.suptitle("Hello")
fig.add_axes([0, 0, 1, 1])
plt.show()
```

```python
fig = plt.figure()
fig.suptitle("Hello")
fig.add_axes([0, 0.1, 0.2, 0.8])
plt.show()
```

```python
fig = plt.figure()
fig.suptitle("Hello")
fig.add_axes([0, 0.1, 0.2, 0.8])
plt.scatter([0.5], [0.5])
plt.show()
```
