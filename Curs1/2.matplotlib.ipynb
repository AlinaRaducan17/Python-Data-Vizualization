# Matplot poate lua si range-uri (ArrayLike)

```python
range_1 = range(4,8)
range_2 = range(6,10)
plt.scatter(range_1, range_2)
plt.show()
```

```python
range_1 = range(4)
range_2 = [3] * 4
plt.scatter(range_1, range_2)
plt.show()
```

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu interesant")
plt.xlabel("Valoare X")
plt.ylabel("Valoare Y")
plt.scatter(x,y);

#pt a salva figura fie cu save as din dreapta, fie
plt.savefig("Figura_mea.png")
```
